```python
import numpy as np 
import matplotlib.pyplot as plt

x=np.arange(1,100)
y=x**2

plt.plot(x,y)
plt.show()
```

```python
# how to style the graphs 
x=np.arange(1,10)
y=x**2

plt.title('Tiltle')
plt.ylabel('x**2')
plt.grid()


# value is already decidede 
plt.xlim(-5,15)
plt.ylim(-10,100)


plt.xlabel('X labels')
plt.plot(x,y ,'r--',label='quadretic ')

# for create multiple cruds in single graph
# line width is used for width controling

plt.plot(x,x+50,'b',label='linear', linewidth=4)

# for display the labels 
plt.legend()

# another labels that you want to display no the graph
plt.text(7.5,50,'Hello')
plt.show()
```

```python
import math
x=np.arange(1,100)
#mapping or loop or (list comprehension is the best methodes in this )
#y=[math.log(i)for i in x]
 
    # list comprehension 
y=list(map(lambda i: math.log(i),x))
plt.plot(x,y)
plt.show()
```

```python
#another way  that is vectorize
import math
x=np.arange(1,100)
# its allow to pass the list on the place of number in log math methode
foo=np.vectorize(math.log)
y=foo(x)
plt.plot(x,y)
plt.show()
```

```python
import math
x=np.arange(1,100)
# its allow to pass the list on the place of number in log math methode
y=np.vectorize(math.log)(x)
# used that in single  line 

plt.plot(x,y)
plt.show()
```

# subplots 

```python
x = np.arange(1,10)
y=x**2

plt.subplot(2,1,1)
    #(211)
plt.plot(x,y)


plt.subplot(2,2,4)
plt.plot(x,y)

plt.subplot(223)    
plt.plot(x,y)

# graph number will left to right and top to bottom 
```
